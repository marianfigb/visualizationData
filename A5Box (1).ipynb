```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('antropometria-dataset-1.csv')
df.head()
```

```python
import seaborn as sns
```

* Question 1 box plot de edad

```python
sns.boxplot(x = df['edad'],     
           ).set_title('Box plot de edad')
plt.show()
```

* Question 2 Box plot de median Peso sexo 2

```python
q2 = pd.pivot_table(df,
                       index = 'desc_ent',
                       values = 'peso',
                        aggfunc = 'median',
                        columns = 'sexo',
                    margins= True
                    
                      
                      )
q2
```

* Question 3 line plot de cadera sexo 1

```python
q3 = pd.pivot_table(df,
                       index = 'desc_ent',
                       values = 'cadera',
                        aggfunc = 'median',
                        columns = 'sexo',
                    margins= True
                    
                      
                      )
q3
```

```python
q3.plot(figsize=(20,10))
```

* Question 4 sistol sexo 1

```python
q4 = pd.pivot_table(df,
                       index = 'desc_ent',
                       values = 'sistol',
                        aggfunc = 'sum',
                        columns = 'sexo',
                      )
q4
```

```python
q4.plot.barh(figsize=(20,10))
```

* Question 5 Histograma Cadera Bins=25

```python
sns.histplot(x = df['cadera'], bins=25,
).set_title('Histplot de cadera')
plt.show()
```
