```python
df['Total'] = df.iloc [:,
                       2:].sum(axis=1)
df.head(2)
```

```python
ic_data = df.loc[['India','China'], '1980':'2013']
ic_data
```

```python
nic = ic_data.transpose() #change axis
nic
```

```python
nic.plot()
```

## Conclusion

# We see a similar migration pattern from Canada and India

```python
nic.plot(figsize=(20,10))
plt.xlabel('Years', fontsize = 17)
plt.xticks(fontsize = 17, rotation =45)
plt.ylabel('Number of Imigrants', fontsize = 17)
plt.yticks(fontsize = 17)
plt.title ('Immigrantion to canada from India and China during 1980 to 2013', fontsize =17)
plt.show()
```
