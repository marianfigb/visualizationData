```python
import seaborn as sns
```

```python
df = pd.read_csv('datasets/Cartwheeldata.csv')
df.head()
```

```python
sns.lmplot( x='Wingspan',
           y='CWDistance',
           data=df,
           #fit_reg = False, #for regresion
           hue = 'Gender', #For any category to plot and change colors
          )
plt.title("Scatter plot of CWDistance vs Wingspan", fontsize =17)
plt.show()
```

```python
sns.swarmplot(x = 'Gender',
             y = 'CWDistance',
             data = df
            )
plt.show()
```

# Box plot of Male data

```python
sns.boxplot(data = df.loc[df['Gender']=='M',
                          ['Age','Height','Wingspan','CWDistance','Score']])
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title("Box plot representation of the Male data", fontsize =15)

plt.show()
```

# 1. Make a comparison of 'Scores' of 'Male' and 'Female' data

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1) #num row, num col, position

sns.boxplot(data = df.loc[df['Gender']=='M',
                          ['Score']
                         ]
           )

ax2 = fig.add_subplot(1,2,2) #num row, num col, position

sns.boxplot(data = df.loc[df['Gender']=='F',
                          ['Score']
                         ]
           )

ax1.title.set_text('Male box plot')
ax2.title.set_text('Female box plot')
plt.show()
```
