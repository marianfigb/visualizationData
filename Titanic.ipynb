```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Titanic.csv") #Abrir csv
df.head()
```

```python
df.rename(columns = {'sex' : 'Sex',
                     'age' : 'Age',
                     'sibsp' : 'Siblings','parch' : 'Parents/Children','fare' : 'Fare','embarked' : 'Embarked','class' : 'Class','who' : 'Category', 'alone' : 'Alone', 'survived' : 'Survived'},
          inplace = True)
df.head()
```

# What is the average age per gender of the people who survived the sinking of the Titanic?

```python
question1 = pd.pivot_table(df,
               index = "Sex",
                values = "Age", 
               aggfunc = "median", 
               columns = "Survived"
              )
question1
```
