# A3 Pivot tables

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("antropometria-dataset-2.csv") #Abrir csv
df.head()
```

* Question 1 Peso promedio de gente de Queretaro

```python
pivot_question1 = pd.pivot_table(df,
               index = 'desc_ent',
               values = 'peso',
              )
pivot_question1
```

* Question 2 Peso median de sexo 2 Yucatan

```python
pivot_question2 = pd.pivot_table(df,
               index = 'desc_ent',
               aggfunc = 'median', 
               values = 'peso', 
               columns = 'sexo'
              )
pivot_question2
```

* Question 3 Datos del peso de sexo 2: Â¿CuÃ¡l es el valor mÃ¡ximo de promedio de los estados analizados? 

```python
pivot_question3 = pd.pivot_table(df,
               index = 'desc_ent',
               aggfunc = 'mean', 
               values = 'peso', 
               columns = 'sexo'
              )
pivot_question3
```

```python
df_adults = pd.read_csv("datasets/adult.csv") #Abrir csv
df_adults.head()
```

* Question 4 Age Divorced HS Grad

```python
pivot_question4 = pd.pivot_table(df_adults,
               index = "marital-status",
                values = "age", 
               aggfunc = "mean", 
               columns = "education"
              )
pivot_question4
```

* Question 5 Median age Male Married AF Spouse Bachelors

```python
pivot_question5 = pd.pivot_table(df_adults,
               index = "marital-status",
                values = "age", 
               aggfunc = "median", 
               columns = ("education","sex"),
              )
pivot_question5
```
