```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Titanic.csv") #Abrir csv
df.head()
```

```python
df.rename(columns = {'sex' : 'Sex',
                     'age' : 'Age',
                     'sibsp' : 'Siblings','parch' : 'Parents/Children','fare' : 'Fare','embarked' : 'Embarked','class' : 'Class','who' : 'Category', 'alone' : 'Alone', 'survived' : 'Survived'},
          inplace = True)
df.head()
```

# What class is more likely to survive the Titanic sinking?

```python
question1 = pd.pivot_table(df,  
                           index='Class',
                           values ="Survived")
question1
```

```python
question1.plot()
```

# How does the gender category assigned affect survival?

```python
question2 = pd.pivot_table(df,
               index = ["Category"],
                values = ["Survived"], 
              )
question2
```

```python
question2.plot()
```

# What is the average fare paid per class?

```python
question3 = pd.pivot_table(df,
               index = "Class",
                values = "Fare", 
               aggfunc = "mean", 
              )
question3
```

```python
question3.plot()
```
