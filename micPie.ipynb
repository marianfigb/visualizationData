```python
df= pd.read_excel('mic2.xlsx', 
                  engine = 'openpyxl')
df
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.style.use('ggplot')
```

```python
counts = df['Age'].value_counts()
```

```python
counts
```

```python
counts = counts.reindex(['<=20','21 - 30', '31 - 40','41 - 50'])
counts
```

```python
type (counts)
counts = counts.to_frame()
type (counts)
```

```python
counts['%'] = (counts.iloc[:,:]/counts.loc[:,'Age'].sum(axis=0))*100
```

```python
counts
```

```python
counts.columns
```

```python
counts=counts.reset_index()
counts
```

```python
counts=counts.rename({'index':'Age_group'},
                    axis = 1)
counts
```

```python
x = list(counts ['Age_group'])
y = list(counts ['Age'])
                 
plt.subplot(1,2,1)
plt.rcParams['figure.figsize'] = (15,7)
plt.bar (x,y,width = 0.7, color = 'darkcyan')

plt.subplot(1,2,2)
explode_list = [0.08,0.09,0.07,0.03]
color_list = ['gold','yellowgreen','coral','skyblue','lightgreen','pink']
counts['Age'].plot(kind='pie',
                  figsize = (15,6),
                  autopct='%1.1f%%',
                  startangle = 90,
                  shadow = True,
                   labels = None,
                   explode = explode_list,
                   colors = color_list,
                  pctdistance = 1.1)
                     
plt.show()
```
