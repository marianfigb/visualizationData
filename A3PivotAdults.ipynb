```python
import pandas as pd
import numpy as np
```

```python
df_adults = pd.read_csv("datasets/adult2.csv") #Abrir csv
df_adults.head()
```

```python
pivot_question4 = pd.pivot_table(df_adults,
               index = "marital-status",
                values = "age", 
               aggfunc = "mean", 
               columns = "education"
              )
pivot_question4

# Promedio age divorced con hs grad?
```

```python
pivot_question5 = pd.pivot_table(df_adults,
               index = "marital-status",
                values = "age", 
               aggfunc = "median", 
               columns = ("education","sex"),
              )
pivot_question5

#median age male married af spouse con bachelors
```

```python
pivot_question5.to_csv('piv.csv')
```
